Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importting the dataset

```python
dataset = pd.read_csv('auto-mpg.csv')
```

Will choose as X the weight of the car and as y the mpg

```python
y = dataset.iloc[:,0].values
```

```python
print(y)
```

```python
X = dataset.iloc[:,4].values
```

```python
print(X)
```

Weight doesnt have missing values

---



Spit Train and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

Training the test set to the model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train.reshape(-1, 1), y_train)
```

Mak

```python
y_pred = regressor.predict(X_test.reshape(-1,1))
```

Visualising the training set results

```python
plt.scatter(X_train, y_train, color = 'red')
# Create a sorted array for plotting the regression line
X_grid = np.arange(min(X_train), max(X_train), 0.1)
X_grid = X_grid.reshape(-1, 1)
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Mpg vs Weight (Training set)')
plt.xlabel('Weight')
plt.ylabel('Mpg')
plt.show()
```

Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'red')
# Create a sorted array for plotting the regression line
X_grid = np.arange(min(X_test), max(X_test), 0.1)
X_grid = X_grid.reshape(-1, 1)
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Mpg vs Weight (Test set)')
plt.xlabel('Weight')
plt.ylabel('Mpg')
plt.show()
```

Model evalutation

```python
print("Intercept:", regressor.intercept_)
print("Slope:", regressor.coef_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = regressor.predict(X_test.reshape(-1, 1))

print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))
```
